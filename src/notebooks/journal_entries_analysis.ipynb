```python
import json
import pandas as pd
from pathlib import Path


base_path = Path.cwd() 
data_path = base_path.parent / "data" / "raw" / "journal_entry.json"


try:
    with open(data_path, 'r') as f:
        data = json.load(f)
    print(f"Success! Found file at: {data_path}")
    df = pd.DataFrame(data)
    display(df.head())
    
except FileNotFoundError:
    print(f"FAILED. Attempted path: {data_path}")
    print(f"Folders visible from here: {[x.name for x in base_path.parent.iterdir()]}")
```
